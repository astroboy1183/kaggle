```python
import pandas as pd
import numpy as np

data = pd.read_csv('train.csv')
```

```python
data.head(5)
```

```python
data = data.drop(['Id'],axis=1)
```

```python
col_wildernessarea = np.array(data.iloc[:,10:14])
col_soiltype = np.array(data.iloc[:,14:-1])
mod_wildernessarea = np.argmax(col_wildernessarea, axis=1)+1
mod_soiltype = np.argmax(col_soiltype,axis=1)+1

# pd.DataFrame(columns=[mod_wildernessarea,mod_soiltype],index=[i for i in range(15120)])
temp = pd.DataFrame(mod_soiltype)
temp.rename(columns={0:'Soil_Type'},inplace=True)
temp1 = pd.DataFrame(mod_wildernessarea)
temp1.rename(columns={0:'Wilderness_Area'},inplace=True)

temp2 = pd.concat([temp,temp1],axis=1)

data = pd.concat([data,temp2],axis=1)
```

```python
# data.iloc[:,10:-3]
data.drop(data.iloc[:,10:-3],axis=1,inplace=True)
```

```python
temp3 = pd.DataFrame(data['Cover_Type'])
data.drop(['Cover_Type'],inplace=True,axis=1)
data = pd.concat([data,temp3],axis=1)
```

```python
data.head(10)
```

```python
from sklearn.preprocessing import StandardScaler

scale = StandardScaler()

# scale.fit(data)
# scale.transform(data)

data_temp = data.drop(['Cover_Type'],axis=1)
data_temp_norm = (data_temp - data_temp.min())/(data_temp.max()-data_temp.min())
data_temp_norm = pd.concat([data_temp_norm,data['Cover_Type']],axis=1)
```

```python
data_temp_norm
```

```python
y = data.iloc[:,-1]
x = data.iloc[:,:-1]
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression

xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.01)

rfc = RandomForestClassifier()

svm = SVC()

lr = LogisticRegression()

rfc.fit(xtrain,ytrain)
svm.fit(xtrain,ytrain)
lr.fit(xtrain,ytrain)
```

```python
testdata = pd.read_csv('test.csv') #read test file
```

```python
testdata = testdata.drop(['Id'],axis=1)
```

```python
from sklearn.metrics import accuracy_score
preds = svm.predict(xtest)

accuracy_score(preds,ytest)
```
